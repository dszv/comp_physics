### Exercise 3.1: Plotting experimental data
##### a) Write a program that reads in the data and makes a graph of sunspots as a function of time.

```python
data1 = loadtxt("sunspots.txt", float)
```

```python
x1a = data1[:,0]
```

```python
y1a = data1[:, 1]
```

```python
plot(x1a, y1a)
xlabel("Months since january 1749")
ylabel("Sunspots")
grid()
show()
```

##### b) Modify your program to display only the first 1000 data points on the graph.

```python
y1b = data1[:1000]
```

```python
plot(y1b[:,0], y1b[:,1] )
xlabel("Months since january 1749")
ylabel("Sunspots")
xlim(-50,1050)
grid()
show()
```

##### c) Modify your program further to calculate and plot the running average of the data, defined by:
$Y_{k} = \frac{1}{2r + 1}\sum^{r}_{m = -r}y_{k+m}$
##### r = 5. Trace el programa con los datos originales y el promedio móvil en el mismo gráfico. Otra vez, los primeros 1000 datos.

```python
r1c = 5
```

```python
def sumayk():
    Y = []
    for i in range(0, 1000):
        suma = 0
        for j in range(-5,6):
            suma = suma + y1a[i+j]
        Y.append( (1/(2*r1c + 1))*(suma) )
    return Y
Y1 = sumayk()
Y1 = array(Y1)
```

```python
plot(y1b[:,0], Y1, "ro", label = r'$Y_{k}$: Running average')
plot(y1b[:,0], y1b[:,1], "b-", label = r'$y_{k}$: Sunspots Numbers')
xlabel("Months since January 1749")
ylabel("Sunspot number")
legend(loc = "best")
grid()
show()
```
