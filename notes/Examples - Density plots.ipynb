#### Example 3.1 - Wave Interference
##### Tenemos: 
##### $\xi_{1}(x,y) = \xi_{0}sin(k r_{1}) \rightarrow r_{1} = \sqrt{(x-x_{1})^{2} + (y-y_{1})^{2}}$
##### $\xi_{2}(x,y) = \xi_{0}sin(k r_{2}) \rightarrow r_{2} = \sqrt{(x-x_{2})^{2} + (y-y_{2})^{2}}$
##### Graficaremos:
##### $\xi(x, y) = \xi_{0}sin(k r_{1}) + \xi_{0}sin(k r_{2})$
##### Consideramos:
##### $\lambda = 5cm$
##### $\xi_{0} = 1cm$
##### $ d = 20 cm$

```python
from math import sqrt,sin,pi
wavelength = 5.0
k = 2*pi/wavelength
xi0 = 1.0
separation = 20.0
side = 100.0
points = 500
spacing = side/points
x1 = side/2 + separation/2
y1 = side/2
x2 = side/2 - separation/2
y2 = side/2
xi = empty([points, points], float)
for i in range(points):
    y = spacing*i
    for j in range(points):
        x = spacing*j
        r1 = sqrt((x-x1)**2 + (y-y1)**2)
        r2 = sqrt((x-x2)**2 + (y-y2)**2)
        xi[i,j] = xi0*sin(k*r1) + xi0*sin(k*r2)
imshow(xi, origin = "lower", extent = [0, side, 0, side])
gray()
title("")
show()
```
