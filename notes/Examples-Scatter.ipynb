#### Example 1: Temperatures and magnitudes of a set of stars.

```python
data = loadtxt("stars.txt", float)
```

```python
scatter(data[:,0], data[:,1])
title('The Hertzsprung-Russell diagram')
xlabel("Temperature")
ylabel("Magnitude")
ylim(-5,20)
xlim(0,13000)
grid()
show()
```

##### Example 2: Inverted axis.
##### The magnitude of a star is defined in such a way that it actually increases as the star gets fainter, so reversing the vertical axis makes sense since it puts the brightest stars at the top.

```python
scatter(data[:,0], data[:,1])
title('Inverted axis on the Hertzsprung-Russell diagram')
xlabel("Temperature")
ylabel("Magnitude")
ylim(20,-5)
xlim(13000,0)
grid()
show()
```
