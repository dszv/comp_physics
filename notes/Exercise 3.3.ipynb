### Excercise 3.3: Mediciones del microscopio de efecto t√∫nel de barrido
Make a density plot of stm.txt that contains a grid of values from scanning tunneling microscope measurements of the (111) surface of silicon. A scanning tunneling microscope (STM) is a device that measures the shape of a surface at the atomic level by tracking a sharp tip over the surface and measuring quantum tunneling current as a function of position.

```python
data3 = loadtxt("stm.txt", float)
```

#### Remind the list of color scheme:
jet: The default heat-map color scheme.
gray: Gray-scale running from black to white.
hot: An alternative heat map that goes black-red-yellow-white.
spectral: A spectrum with 7 clearly defined colors, plus black and white.
bone: An alternative gray-scale with a hint of blue.
hsv: A rainbow scheme that starts and ends with red

```python
imshow(data3, origin = "lower", extent = [-300, 300, -300, 300])
title("Surface of silicon (111) displayed by a STM")
bone()
show()
```
