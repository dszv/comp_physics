# Problem 3

```python
import vpython as vp
```

```python
scene1 = vp.canvas()

L  = 1        # number of atoms = 2*L+1
R1 = 0.227    # radius of Na atom
R2 = 0.175    # radius of Cl atom

for i in range(-L,L+1,1):
    for j in range(-L,L+1,1):
        for k in range(-L,L+1,1):
            if (i+j+k) % 2 == 0:
                vp.sphere(pos=vp.vector(i,j,k), radius=R2)
            else:
                vp.sphere(pos=vp.vector(i,j,k), radius=R1, color=vp.color.red)
```

# Problem 3b

```python
scene2 = vp.canvas()

L = 1        # number of atoms = 2*L+1
R = 0.1      # radius of the atom
o = 0.5      # offset

for i in range(-L,L+1,1):
    for j in range(-L,L+1,1):
        for k in range(-L,L+1,1):
            vp.sphere(pos=vp.vector(i,j,k), radius=R)
            if i != L and j != L:
                vp.sphere(pos=vp.vector(i+o,j+o,k), radius=R, color=vp.color.green)
                vp.sphere(pos=vp.vector(i+o,k,j+o), radius=R, color=vp.color.green)
                vp.sphere(pos=vp.vector(k,i+o,j+o), radius=R, color=vp.color.green)
```

# Problem 4a

## Equations of motion

\begin{align}
    x(t) & = R \cos \left( 2 \pi t/T \right) \\
    y(t) & = R \sin \left( 2 \pi t/T \right)
\end{align}

# Problem 4b

```python
# Data
c1 = 1.5*0.05
planets = ['Sun','Mercury','Venus','Earth','Mars','Jupyter','Saturn']
radii   = c1*np.array([69.57, 2.44, 6.052, 6.371, 3.386, 69.173, 57.316])
orbit   = c1*np.array([0.01, 57.9, 108.2, 149.6, 227.9, 778.5, 1433.4])
period  = np.array([0, 88.0, 224.7, 365.3, 687.0, 4331.6, 10759.2])

omega = c1*2*np.pi/orbit

scene3 = vp.canvas()

for i in range(len(planets)):
    planets[i] = vp.sphere(radius=radii[i])
    
for t in np.arange(0,100,0.1):
    vp.rate(35)
    for i in range(len(planets)):
        x = orbit[i]*np.cos(omega[i]*t)
        y = orbit[i]*np.sin(omega[i]*t)
        planets[i].pos = vp.vector(x,y,0)
```
